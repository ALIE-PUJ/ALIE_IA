```python
import requests
import pandas as pd
import random
import time

# Mezclar preguntas
random.shuffle(alie_questions)

# Inicializar lista de respuestas
responses = []

# Función para hacer la solicitud y almacenar la respuesta
def get_response(pregunta):
    start_time = time.time()
    response = requests.post(
        "http://localhost:3000/alie",
        json={"input": pregunta, "priority": "False"},
        headers={'Content-Type': 'application/json; charset=utf-8'}
    )
    end_time = time.time()
    response_time = end_time - start_time
    
    # Decodificar la respuesta
    response_data = response.json()
    answer = response_data.get("answer", "No se recibió respuesta")
    
    return answer, response_time

# Obtener respuestas para todas las preguntas
for item in alie_questions:
    pregunta = item["pregunta"]
    respuesta, tiempo = get_response(pregunta)
    responses.append({
        "Pregunta": pregunta,
        "Respuesta": respuesta,
        "Supervisión": None,
        "Tiempo de Respuesta (s)": round(tiempo, 2)
    })
```

```python
# Convertir a DataFrame
df_responses = pd.DataFrame(responses)

# Mostrar el DataFrame
display(df_responses)
```

```python
import pandas as pd
from flask import Flask, render_template_string, request

# Crear una aplicación Flask
app = Flask(__name__)

@app.route('/', methods=['GET', 'POST'])
def home():
    global df_responses
    if request.method == 'POST':
        evaluations = []
        for idx in range(len(df_responses)):
            evaluation = request.form.get(f'evaluation{idx}')
            evaluations.append(evaluation)

        if None in evaluations:
            return "Error: No todas las respuestas fueron evaluadas.", 400

        correct_count = evaluations.count('correct')
        total_count = len(evaluations)
        effectiveness = (correct_count / total_count) * 100 if total_count > 0 else 0
        fallback = (total_count - correct_count) / total_count * 100 if total_count > 0 else 0
        df_responses['Supervisión'] = evaluations

        table_html = df_responses.to_html(classes='table table-striped', index=False, escape=False)

        html = f'''
        <html>
        <head>
            <title>Evaluación de Respuestas</title>
            <link rel="stylesheet" href="https://stackpath.bootstrapcdn.com/bootstrap/4.5.2/css/bootstrap.min.css">
            <style>
                .table-container {{
                    max-height: auto;
                    overflow-y: auto;
                    overflow-x: hidden;
                }}
                .table {{
                    width: 100%;
                    table-layout: auto;
                }}
                th, td {{
                    overflow: hidden;
                    text-overflow: ellipsis;
                    word-wrap: break-word;
                }}
                td {{
                    max-width: 300px;
                }}
            </style>
        </head>
        <body>
            <div class="container">
                <h1 style="margin-top: 20px;">Evaluación de Respuestas</h1>
                <div class="table-container">
                    {table_html}
                </div>
                <h2>Resultados de Evaluación</h2>
                <p>Cantidad de Preguntas: {len(alie_questions)}</p>
                <p>Índice de Efectividad: {effectiveness:.2f}%</p>
                <p>Índice de Fallback: {fallback:.2f}%</p>
                <a href="/">Volver a evaluar</a>
            </div>
        </body>
        </html>
        '''
        return render_template_string(html)

    table_html = df_responses.to_html(classes='table table-striped', index=False, escape=False)

    evaluation_inputs = ''.join(
        f'''
        <tr>
            <td>{idx + 1}</td>
            <td>{row['Pregunta']}</td>
            <td>{row['Respuesta']}</td>
            <td>
                <input type="radio" name="evaluation{idx}" value="correct" required> Correcto
                <input type="radio" name="evaluation{idx}" value="incorrect"> Incorrecto
            </td>
            <td>{row['Tiempo de Respuesta (s)']}</td>
        </tr>
        '''
        for idx, row in df_responses.iterrows()
    )

    html = f'''
    <html>
    <head>
        <title>Evaluación de Respuestas</title>
        <link rel="stylesheet" href="https://stackpath.bootstrapcdn.com/bootstrap/4.5.2/css/bootstrap.min.css">
        <style>
            .table-container {{
                max-height: auto;
                overflow-y: auto;
                overflow-x: hidden;
            }}
            .table {{
                width: 100%;
                table-layout: auto;
            }}
            th, td {{
                overflow: hidden;
                text-overflow: ellipsis;
                word-wrap: break-word;
            }}
            td {{
                max-width: 300px;
            }}
        </style>
    </head>
    <body>
        <div class="container">
            <h1 style="margin-top: 20px;">Evaluación de Respuestas</h1>
            <form method="POST">
                <div class="table-container">
                    <table class="table table-striped">
                        <thead>
                            <tr>
                                <th>Número</th>
                                <th>Pregunta</th>
                                <th>Respuesta</th>
                                <th>Supervisión</th>
                                <th>Tiempo de Respuesta (s)</th>
                            </tr>
                        </thead>
                        <tbody>
                            {evaluation_inputs}
                        </tbody>
                    </table>
                </div>
                <button type="submit" class="btn btn-primary" style="margin-top: 8px;">Enviar Evaluación</button>
            </form>
        </div>
    </body>
    </html>
    '''
    return render_template_string(html)

if __name__ == '__main__':
    app.run(port=7000, debug=False)
```
